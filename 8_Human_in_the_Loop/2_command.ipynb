```python
# Import StateGraph to create workflow graph
# START and END represent beginning and ending of graph
from langgraph.graph import StateGraph, START, END

# Import Command to control graph flow dynamically
from langgraph.types import Command

# Import TypedDict to define state structure
from typing import TypedDict


# Define the structure of graph state
class State(TypedDict):

    # State contains one key called text
    # It stores string data
    text: str


# First node function
def node_a(state: State):

    # Print which node is running
    print("Node A")

    # Return command object
    return Command(

        # Move next to node_b
        goto="node_b",

        # Update state
        update={

            # Add "a" to existing text
            "text": state["text"] + "a"
        }
    )


# Second node function
def node_b(state: State):

    # Print current node
    print("Node B")

    # Return command object
    return Command(

        # Move to node_c
        goto="node_c",

        # Update state
        update={

            # Add "b" to text
            "text": state["text"] + "b"
        }
    )


# Third node function
def node_c(state: State):

    # Print current node
    print("Node C")

    # Return command object
    return Command(

        # End graph execution
        goto=END,

        # Update final state
        update={

            # Add "c" to text
            "text": state["text"] + "c"
        }
    )


# Create graph object using State
graph = StateGraph(State)


# Add node_a function to graph
graph.add_node("node_a", node_a)

# Add node_b function to graph
graph.add_node("node_b", node_b)

# Add node_c function to graph
graph.add_node("node_c", node_c)


# Set first node where execution starts
graph.set_entry_point("node_a")


# Compile graph into executable app
app = graph.compile()


# Run graph with initial state
response = app.invoke({

    # Initial text is empty
    "text": ""
})


# Print final output
response
```
